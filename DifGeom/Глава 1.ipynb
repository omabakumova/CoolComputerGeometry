```python


def plot_vector_function(r_function, t_range, title="  "):
    """
        r(t).

    
        r_function: ,    t    r(t).
        t_range:    t.
             ,     ,   ,   `t` .

       :
    
     `r(t)`:    `r(t)`           `t`.
     `t`:   `t` ,  `r(t)`  ,     .
    :        ,  `t` . 

    """

    #   t   
    t = np.linspace(t_range[0], t_range[1], 100)

    #    
    x = [r_function(t_i)[0] for t_i in t]
    y = [r_function(t_i)[1] for t_i in t]
    z = [r_function(t_i)[2] for t_i in t]

    #   
    dx = [-np.sin(t_i) for t_i in t]
    dy = [np.cos(t_i) for t_i in t]
    dz = [1 for _ in t]

    #  3D 
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x, y, z, label='')

    #  ,   
    ax.quiver(x, y, z, dx, dy, dz, length=0.5, normalize=True, color='r')

    #  
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title(title)
    ax.legend()
    plt.show()

#  r(t) = (cos(t), sin(t), t)
def r(t):
    return np.array([np.cos(t), np.sin(t), t])

#    t
t_range = (0, 6 * np.pi)  #     

#  
plot_vector_function(r, t_range, "  r(t) = (cos(t), sin(t), t)")
```
